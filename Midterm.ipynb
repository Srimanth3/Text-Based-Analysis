## **Making a PipeLine**

```python
# making a “pipeline"
pipeline = [str.lower, tokenize]
def prepare(text, pipeline):
    tokens = text
    for fun in pipeline:
        tokens = fun(tokens)
    return tokens
```

```python
# Runing the pipeline
df['tokens'] = df['CONTENT'].apply(prepare, pipeline=pipeline)
```

```python
#Clean the text
#Create a function to clean the Contents
def cleanTxt(CONTENT):
    CONTENT = re.sub(r'@[A-Za-z0-9]+', '', CONTENT) #Remove @mentions replace with blank
    CONTENT = re.sub(r'#', '', CONTENT) #Remove the '#' symbol, replace with blank
    CONTENT = re.sub(r'RT[\s]+', '', CONTENT) #Removing RT, replace with blank
    CONTENT = re.sub(r'https?:\/\/\S+', '', CONTENT) #Remove the hyperlinks
    CONTENT= re.sub(r':', '', CONTENT) # Remove :
    
    return CONTENT
#Cleaning the CONTENT
df["CONTENT"]=df["CONTENT"].apply(cleanTxt)

#Show the clean CONTENT
df.head()
```

## **remove emoji & Unicode from the Content Column**

```python
#Next we have to remove emoji & Unicode from the Content Column.
def remove_emoji(string):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)

#Cleaning the text
df["CONTENT"]=df["CONTENT"].apply(remove_emoji)

#Show the clean text
df.head()
```

## **Getting the Bing Liu lexicon**

```python
from sklearn import preprocessing # Importing Preprocessing
import nltk                       # Importing nltk
nltk.download('opinion_lexicon')
from nltk.corpus import opinion_lexicon # Importing oponion_Lexicon
from nltk.tokenize import word_tokenize # Imorting word_tokenize

print('Total number of words in opinion lexicon', len(opinion_lexicon.words()))
print('Examples of positive words in opinion lexicon',
      opinion_lexicon.positive()[:10])
print('Examples of negative words in opinion lexicon',
      opinion_lexicon.negative()[:10])
```

## **Creating the Dictionary**

```python
nltk.download('punkt')
df.rename(columns={"reviewText": "text"}, inplace=True)
pos_score = 1
neg_score = -1
word_dict = {}
 
# Adding the positive words to the dictionary
for word in opinion_lexicon.positive():
        word_dict[word] = pos_score
      
# Adding the negative words to the dictionary
for word in opinion_lexicon.negative():
        word_dict[word] = neg_score
```

## **Math Function**

```python
def bing_liu_score(text): 
    sentiment_score = 0
    bag_of_words = word_tokenize(text.lower())
    for word in bag_of_words:
        if word in word_dict:
            sentiment_score += word_dict[word]
    return sentiment_score
```

## **OutPut Bing_Liu_Score**

```python
df['CONTENT'].fillna('no review', inplace=True)
df['Bing_Liu_Score'] = df['CONTENT'].apply(bing_liu_score)
```
